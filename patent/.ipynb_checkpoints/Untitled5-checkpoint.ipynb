```python
# define regex patterns

company_alias = ["회사", "기업", "브랜드", "브랜드명", "업체", "개발사", "제조사", "제작사", "제약사", "대기업", "후발업체", "후발 업체"]
patterns = [r'(%s){1}인{0,1} ㈜{0,1}([ㄱ-ㅣ가-힣()a-zA-Z0-9.%%\-\'"]+)에서{1}는{0,1} ' % '|'.join(company_alias),
            r'(%s){1}인{0,1} ㈜{0,1}([ㄱ-ㅣ가-힣()a-zA-Z0-9.%%\-]+)[은는가와과도의,]{1} ' % '|'.join(company_alias),
            r' ([ㄱ-ㅣ가-힣()a-zA-Z㈜.]+)[이가] 개발한 ',
            r' ([ㄱ-ㅣ가-힣()a-zA-Z㈜.]+)의 기술은 ',
            r' ([ㄱ-ㅣ가-힣()a-zA-Z㈜.]+)와 같은 기업들{0,1}은 ',
            r'(([ㄱ-ㅣ가-힣()a-zA-Z0-9.%%\-\'"]+[,|·] {0,1})*[ㄱ-ㅣ가-힣()a-zA-Z0-9.%%\-\'"]+) 등 .{0,10}(%s){1}'%'|'.join(company_alias),
            r'설립된 ([ㄱ-ㅣ가-힣()a-zA-Z0-9.%%\-\""]+)[은는가와과도의등,]']
            # r'([ㄱ-ㅣ가-힣()a-zA-Z0-9.%%\-\""]+)\s([ㄱ-ㅣ가-힣()a-zA-Z0-9.%%\-\""]+)[(（]www\.\w{1,}\..{3,5}[/)）]']

except_company_list = ["입장", "입장에서", "경험을","경험", "본인", \
                       "자신","관계자", "중에", "본사", "자사", "배송기사", \
                       "제품", "측", "또", "네트워크"]
```

```python
for i, url in enumerate(html_list):
    print("URL", str(i) + ". " + url)
    html_soup, html_text, footer_soup, footer_text, title = parse_html(url)
    if html_text == None: continue
    for j, pattern in enumerate(patterns):
        if j is not 6 : continue
        res = html_soup.findAll(text=re.compile(pattern))
        print("pattern", j)
        for k, item in enumerate(res):
            print(str(k) + ")")
            cor_word_area = item.strip()
            if len(re.findall(pattern, cor_word_area)) == 0 :  " " + cor_word_area + " "
            cor_word = re.findall(pattern, cor_word_area)[0]
            if j in [0, 1]:
                cor_word = [cor_word[1]]
            elif j == 5 :
                cor_word = cor_word[0]
                if "," in cor_word : cor_word = [e.strip() for e in cor_word.split(",")]
                elif "·" in cor_word : cor_word = [e.strip() for e in cor_word.split("·")]
                else : cor_word = [cor_word.strip()]
                    
            if type(cor_word) == str : cor_word = [cor_word]
    
            for g, each in enumerate(cor_word):
                if each in except_company_list : continue
                # remove all strings in parenthesis with it
                # print(cor_word)
                each = re.sub(r'\([^)]*\)', '', each)
                print(str(g) + "." + each)
```
