```python
def identify_file_format(url):
    file_format = ""
    res = urllib.request.urlopen(url)
    content_type, _ = cgi.parse_header(res.headers.get("Content-type", ""))
    if (".pdf" in url) | (content_type == "application/pdf"):
        file_format = "pdf"
    elif ".hwp" in url:
        file_format = "hwp"
    elif ".docx" in url:
        file_format = "docx"
    else:
        _, params = cgi.parse_header(res.headers.get("Content-Disposition", ""))
        if "filename" in params :
            filename = params["filename"]
            if ".pdf" in filename:
                file_format = "pdf"
            elif ".hwp" in filename:
                file_format = "hwp"
            elif ".docx" in filename:
                file_format = "docx"

    return file_format, filename
```

```python
def identify_file_name(url):
    res = urllib.request.urlopen(url)
    _, params = cgi.parse_header(res.headers.get("Content-Disposition", ""))
    print(params)
    if "filename" in params:
        filename = params["filename"]
    else :
        filename = ""
        
    return filename
```
